```python
def varyansKesikliOlasilik():
  global X,PX
  ex=EX()
  print("E[X]:",ex)
  ex2=EX2()
  print("E[X^2]:",ex2)
  var=ex2-pow(ex,2)
  print("Var[X]:",var)

def EX():
  global X,PX
  ex=0
  for i in range(len(X)):
    ex+=X[i]*PX[i]
  return ex

def EX2():
  global X,PX
  ex2=0
  for i in range(len(X)):
    ex2+=pow(X[i],2)*PX[i]
  return ex2
```

```python
X=[1,2,3,4,5,6,7,8]
PX=[0.02,0.08,0.15,0.19,0.24,0.17,0.10,0.04,0.01]
varyansKesikliOlasilik()
```
