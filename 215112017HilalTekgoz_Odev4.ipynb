**Soru :** Bir çift zar atılıyor. Üst yüze gelenlerden birinin 5 olduğu ***bilindiğine göre,*** üst yüzlere gelen sayıların toplamının 8 olma olasılığı kaçtır?


**Cevap :** 

```
* Tüm olası sonuçlarımız 36'dır. 

* Üst yüze gelenlerden birinin 5 olduğu durumlar : {5,1} {5,2} {5,3} {5,4} {5,5} {5,6}
                                                   {1,5} {2,5} {3,5} {4,5}       {6,5}
                                                                                Toplam 11 olası sonucumuz vardır. 

* Üst yüzlere gelen sayıların toplamının 8 olduğu durumlar: {5,3} {3,5}

* Sonuç : 2/11 = 0.18


```




````python
a = ['{5,1}', '{5,2}', '{5,3}', '{5,4}', '{5,5}', '{5,6}', '{1,5}', '{2,5}','{3,5}', '{4,5}' ,'{6,5}']

b = ['{5,3}','{3,5}']

result = len(b)/len(a)
result
````
